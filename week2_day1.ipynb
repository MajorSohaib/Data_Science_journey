```python
marks = np.array([
    [56, 34, 77],
    [74, 33, 15],
    [88, 25, 56],
    [65, 24, 99]
])

print("full array: ", marks)
print("Third student marks: ", marks[2])
print("Second subject marks of all students: ", marks[:, 1])
print("Marks of student 2 subject 3: ", marks[1, 2])
```

```python
import pandas as pd
import numpy as np

marks = np.array([
    [56, 34, 77],
    [74, 33, 15],
    [88, 25, 56],
    [65, 24, 99]
])

df_marks = pd.DataFrame(
    marks,
    index=['Ali', 'Sara', 'Sohaib', 'Ahmed'],
    columns=['Science', 'Maths', 'Urdu']
)

print(df_marks)
```

```python
marks = np.array([
    [56, 34, 77],
    [74, 33, 15],
    [88, 25, 56],
    [65, 24, 99]
])

print("Average per student: ", np.mean(marks, axis=1))
print("Average per subject: ", np.mean(marks, axis=0))
print("Average marks: ", np.max(marks))
print("Highest per subject: ", np.max(marks, axis=0))
```
