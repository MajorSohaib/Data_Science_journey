```python
products = [
    {"name": "Laptop", "price": 150000, "sold": 5},
    {"name": "Phone", "price": 80000, "sold": 12},
    {"name": "Tablet", "price": 60000, "sold": 8},
    {"name": "Headphones", "price": 15000, "sold": 20},
    {"name": "Charger", "price": 5000, "sold": 35},
]

#Print every product name and price
for product in products:
    print(product["name"] , product["price"])


print( ".\n")

#Calculate and print total revenue per product (price × sold)

for product in products:
    revenue = product["price"] * product["sold"]
    print(product["name"], "revenue: ", revenue )

print(".\n")



#Find and print the most sold product

most_sold = products[0]

for product in products:
    if product["sold"] > most_sold["sold"]:
        most_sold = product

print("Most Sold Product:", most_sold["name"])
print("Units Sold:", most_sold["sold"])
print(".\n")

#Print only products that cost more than 50000
print("Product above 50000: ")
for product in products:
    if product["price"] > 50000:
        print(product["name"])
```
