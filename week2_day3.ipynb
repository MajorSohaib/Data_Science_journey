```python
import pandas as pd
import numpy as np

url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
df = pd.read_csv(url)

print("Loaded " , df.shape)
print(df.head())
print (df.isnull().sum())
print(df.groupby('Sex')['Survived'].mean())
```

# Apply()

```python
import pandas as pd
import numpy as np

def categorize_age(age):
  if pd.isna(age):
    return 'Unknown'
  elif age < 18:
    return 'Child'
  elif age < 35:
    return 'Young Adult'
  elif age < 60:
    return 'Adult'
  else:
    return 'Senior'

df['AgeCategory'] = df['Age'].apply(categorize_age)

print(df[['Age', 'AgeCategory']].head(10))
```

# **Merge()**

```python
students = pd.DataFrame({
    'StudentID': [1, 2, 3, 4],
    'Name' : ['Ahmed', 'Ali', 'Sohaib', 'Usama']
})
marks = pd.DataFrame({
    'StudentID': [1, 2, 3, 4],
    'Marks': [87, 44, 88, 23]

})

print("Student table: ")
print(students)
print("\nMarks table: ")
print(marks)

combined = pd.merge(students, marks, on='StudentID')
print("\nCombined Table:" )
print(combined)
```

```python
students = pd.DataFrame({
    'StudentID': [1, 2, 3, 4, 5],
    'Name': ['Ali', 'Sara', 'Umar', 'Hina', 'Bilal']
})

marks = pd.DataFrame({
    'StudentID': [1, 2, 3, 4],
    'Marks': [78, 45, 91, 33]
})

# Inner merge - only matching rows
inner = pd.merge(students, marks, on='StudentID', how='inner')
print("Inner merge:")
print(inner)

# Left merge - keep all students even without marks
left = pd.merge(students, marks, on='StudentID', how='left')
print("\nLeft merge:")
print(left)
```

```python
pivot = pd.pivot_table(
    df,
    values = 'Survived',
    index = 'Sex',
    columns = 'Pclass',
    aggfunc = 'mean'
)
print(pivot)
```
