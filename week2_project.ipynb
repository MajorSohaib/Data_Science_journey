```python
import pandas as pd
import numpy as np

print("I am doing it done today!!!")

url = "https://raw.githubusercontent.com/ageron/handson-ml2/master/datasets/housing/housing.csv"
df =  pd.read_csv(url)
print("Loaded! ", df.shape)
print(df.head())
```

**What is the average house value by ocean proximity?**

```python
df.groupby('ocean_proximity')['median_house_value'].mean()
```

**Which areas have the highest median income?**

```python
df.groupby('ocean_proximity')['median_income'].mean()
```

 we can see that the houses that are 1H OCEAN has the highest median income.

**Question 3**

Is there a relationship between income and house value?

```python
df['median_income'].corr(df['median_house_value'])
```

There is a strong positive correlation of 0.69 between median income and house value. This means as income increases, house value tends to increase as well. 69% of the variation in house prices can be explained by income levels.

**Question 4:**

"What is the distribution of house ages?"

```python
import matplotlib.pyplot as plt
df['housing_median_age'].plot(kind='hist', bins=20)
plt.title('Distribution of House Ages')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

Most houses in California are between 30-40 years old,
suggesting a large construction boom occurred 30-40 years ago.

**Chart 1 — Average house value by ocean proximity:**

```python
df.groupby('ocean_proximity')['median_house_value'].mean().plot(kind='bar', color='steelblue')
plt.title('Average house value by ocean proximity')
plt.xlabel('Ocean proximity')
plt.ylabel('Average house value')
plt.xticks(rotation=45)
plt.show()
```

**Chart 2 — Number of houses by ocean proximity**

```python
df.groupby('ocean_proximity')['median_house_value'].value_counts().plot(kind='bar', color='orange')
plt.title('Number of houses by ocean proximity')
plt.xlabel('Ocean Proximity')
plt.ylabel('Number of houses')
plt.xticks(rotation=45)
plt.show()
```
