```python
import pandas as pd
url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
df = pd.read_csv(url)
print("loaded", df.shape)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Survival rate

df['Survived'].value_counts().plot(kind='bar' , color=['grey' , ''])
plt.title('Survival Count')
plt.xlabel('Survived (0=Died , 1=Survived)')
plt.ylabel('No of passengers')
plt.show()
```

```python
#Survival rate by gender

df.groupby('Sex')['Survived'].mean().plot(kind='bar' , color=['pink' , 'blue'])
plt.title('Survival rate by Gender')
plt.xlabel('Gender')
plt.ylabel('Survival rate')
plt.show()
```

```python
df.groupby('Pclass')['Survived'].mean().plot(kind='bar' , color=['gold' , 'silver' , 'brown'])
plt.title('Survival rate by Passenger class')
plt.xlabel('Passenger class')
plt.ylabel('Survival rate')
plt.show()
```

```python
df['Survived'].value_counts().plot(kind='pie', labels=['Died' , 'Survived'], color=['Green', 'red'], autopct= '%1.1f%%')
plt.title('Survival Distribution')
plt.ylabel('')
plt.show()
```
