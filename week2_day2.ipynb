Array reshapping


```python
arr = np.array([1, 2, 3, 4, 5, 6])
print("Original shape: ", arr.shape)

arr_2d = arr.reshape(2, 3)
print("Reshaped array of 2x3 is: ")
print(arr_2d)

arr_3x2 = arr.reshape(3, 2)
print("Reshaped array of 3x2 is: ")
print(arr_3x2)
```

Autoshape





```python
arr = np.array([1,2, 3, 4, 5, 6])

auto_reshape = arr.reshape(2, -1)
print("The autoreshape is: ")
print(auto_reshape)
print(arr.shape)
print("Shape: ", auto_reshape.shape)
```

```python
import pandas as pd
import numpy as np

url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
df = pd.read_csv(url)

ages = df['Age'].dropna().to_numpy()

print("Type: ", type(ages))
print("Shape: ", ages.shape )
print("Mean age: ", round(np.mean(ages), 2))
print("Std of age: ", round(np.std(ages), 2))
print("Min age:", np.min(ages))
print("Max age: ", np.max(ages))
```
