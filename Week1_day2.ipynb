```python
#Temps in celcius

temps = [22, 35, 18, 40, 28, 15, 33]

fahrenheit = []
for t in temps:
  f = t * 9/5 +32
  fahrenheit.append(f)

print("Temperatures in Celcius: ",fahrenheit)

print("Temperatures above 30°C:")
for t in temps:
  if t > 30:
    print(t)
```

```python
#Dictionaries and lists

#a) Print names of all Engineering employees
employees = [
    {"name": "Ahmed", "department": "Engineering", "salary": 85000},
    {"name": "Fatima", "department": "Marketing", "salary": 62000},
    {"name": "Hassan", "department": "Engineering", "salary": 91000},
    {"name": "Zara", "department": "HR", "salary": 58000},
    {"name": "Omar", "department": "Engineering", "salary": 78000},
]

for emp in employees:
  if emp["department"] == "Engineering":
    print(emp["name"])


#b) Calculate average salary of Engineering department

total_salary = 0
count = 0

for emp  in employees:
  if emp["department"] == "Engineering":
     total_salary += emp["salary"]
     count += 1

average = total_salary/count
print("Average salary of Enginnering departmenty: ", average)


#c) Find highest paid employee



```
